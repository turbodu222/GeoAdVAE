# Plot Morphology

```python
# -*- coding: utf-8 -*-
"""
Batch plot SWC neurons (XY projection) in a single figure with separate subplots.
- Reads neuron_{i}.swc for i in [START_IDX, END_IDX]
- Each neuron is drawn in its own axis (no overlay)
- Colors: soma=black dot, basal(type=3)=blue, apical(type=4)=red, others=gray
"""

import os
import math
import numpy as np
import matplotlib.pyplot as plt

# -------------------------- Config (edit these) -------------------------- #
FOLDER = "/Users/apple/Desktop/KLin_Group/Project_2024/data/Morpho_data/dataset/Simulation/v5/swc_files"
START_IDX = 1     # neuron_a.swc -> a
END_IDX   = 10    # neuron_b.swc -> b
SAVE_PATH = os.path.join(FOLDER, f"neurons_{START_IDX}_{END_IDX}.png")
MAX_COLS  = 5     # max number of columns in the grid
DPI       = 200
# ------------------------------------------------------------------------ #

def load_swc(filepath):
    """Parse an SWC file into a list of nodes (dicts).
    SWC columns: id, type, x, y, z, radius, parent
    Lines starting with '#' are comments and will be ignored.
    """
    nodes = []
    with open(filepath, "r") as f:
        for line in f:
            line = line.strip()
            if not line or line.startswith("#"):
                continue
            parts = line.split()
            if len(parts) < 7:
                continue
            try:
                nid    = int(float(parts[0]))
                ntype  = int(float(parts[1]))
                x      = float(parts[2])
                y      = float(parts[3])
                z      = float(parts[4])
                r      = float(parts[5])
                parent = int(float(parts[6]))
                nodes.append({
                    "id": nid, "type": ntype, "x": x, "y": y, "z": z,
                    "r": r, "parent": parent
                })
            except ValueError:
                # Skip malformed lines safely
                continue
    return nodes

def plot_neuron_xy(ax, nodes, title=None, show_legend=False):
    """Plot one neuron (XY projection) into the provided axis."""
    if not nodes:
        ax.set_title(f"{title or 'Empty'} (no nodes)")
        ax.axis("off")
        return

    # Build a lookup from node id -> (x, y, type)
    pos = {n["id"]: (n["x"], n["y"]) for n in nodes}
    typ = {n["id"]: n["type"] for n in nodes}

    # Draw edges by parent links
    for n in nodes:
        pid = n["parent"]
        if pid <= 0 or pid not in pos:
            continue
        (x1, y1) = pos[pid]
        (x2, y2) = pos[n["id"]]
        if n["type"] == 3:       # basal
            ax.plot([x1, x2], [y1, y2], 'b-', alpha=0.5, linewidth=0.6)
        elif n["type"] == 4:     # apical
            ax.plot([x1, x2], [y1, y2], 'r-', alpha=0.5, linewidth=0.6)
        else:                     # other
            ax.plot([x1, x2], [y1, y2], '-',  color='0.6', alpha=0.5, linewidth=0.5)

    # Soma as a black dot (type==1). Some SWCs may have multiple type=1 entries; plot them all.
    soma_xy = [(n["x"], n["y"]) for n in nodes if n["type"] == 1]
    if soma_xy:
        xs, ys = zip(*soma_xy)
        ax.scatter(xs, ys, c='black', s=15, marker='o', zorder=3, label='Soma')

    # Autoscale with padding
    xs = [n["x"] for n in nodes]
    ys = [n["y"] for n in nodes]
    x_min, x_max = min(xs), max(xs)
    y_min, y_max = min(ys), max(ys)
    x_rng = max(x_max - x_min, 10.0)
    y_rng = max(y_max - y_min, 10.0)
    x_pad = 0.1 * x_rng
    y_pad = 0.1 * y_rng
    x_ctr = 0.5 * (x_min + x_max)
    y_ctr = 0.5 * (y_min + y_max)
    ax.set_xlim(x_ctr - x_rng/2 - x_pad, x_ctr + x_rng/2 + x_pad)
    ax.set_ylim(y_ctr - y_rng/2 - y_pad, y_ctr + y_rng/2 + y_pad)

    ax.set_aspect('equal', adjustable='box')
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    if title:
        ax.set_title(title, fontsize=10)

    if show_legend:
        # Dummy handles for legend
        ax.plot([], [], 'r-', alpha=0.6, label='Apical (type=4)')
        ax.plot([], [], 'b-', alpha=0.6, label='Basal (type=3)')
        ax.legend(loc='upper right', fontsize=8, frameon=False)

def visualize_swcs(folder, start_idx, end_idx, save_path=None, max_cols=5, dpi=150):
    """Create a figure with subplots for neuron_{i}.swc where i in [start_idx, end_idx]."""
    filepaths = []
    for i in range(start_idx, end_idx + 1):
        fp = os.path.join(folder, f"neuron_{i}.swc")
        if os.path.exists(fp):
            filepaths.append(fp)
        else:
            print(f"[WARN] Missing file: {fp} (skipped)")

    if not filepaths:
        raise FileNotFoundError("No SWC files found in the given index range.")

    n = len(filepaths)
    cols = min(max_cols, n)
    rows = math.ceil(n / cols)

    fig, axes = plt.subplots(rows, cols, figsize=(cols * 3.3, rows * 3.3), dpi=dpi)
    if isinstance(axes, np.ndarray):
        axes = axes.flatten()
    else:
        axes = np.array([axes])

    for idx, (ax, fp) in enumerate(zip(axes, filepaths)):
        nodes = load_swc(fp)
        title = os.path.basename(fp)
        plot_neuron_xy(ax, nodes, title=title, show_legend=(idx == 0))

    # Hide any unused axes
    for j in range(len(filepaths), len(axes)):
        axes[j].axis("off")

    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, bbox_inches='tight')
        print(f"[OK] Figure saved to: {save_path}")
    else:
        plt.show()
    plt.close(fig)

if __name__ == "__main__":
    visualize_swcs(FOLDER, START_IDX, END_IDX, save_path=SAVE_PATH, max_cols=MAX_COLS, dpi=DPI)
```
